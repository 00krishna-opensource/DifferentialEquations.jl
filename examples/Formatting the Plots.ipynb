# Formatting The Plots

Since the plotting functionality is implemented as a recipe to Plots.jl, [all of the options open to Plots.jl can be used in our plots](https://juliaplots.github.io/supported/). Let's first get the solution to some SDE. Here I will use one of the example stochastic differential equations from `/premades/premade_problems.jl`. `waveSDEExample()` will give me a stochastic-periodic solution. As with all commands in DifferentialEquations.jl, I got a plot of the solution by calling `solve` on the problem, and `plot` on the solution:

```julia
using DifferentialEquations
prob = waveSDEExample()
Δt = 1//2^(4) #The initial timestepping size. It will automatically assigned if not given.
tspan = [0,1] # The timespan. This is the default if not given.
sol =solve(prob::SDEProblem,tspan,Δt=Δt,save_timeseries=true,alg=:EM)
plot(sol,plot_analytic=true)
```

Here I used a non-standard keyword argument, `plot_analytic`, to add the true solution to the plot (which is able to be done because the true solution is defined in `prob`). However, we can add any of the commands from the Plots.jl documentation. For example, we can change the linewidths with `w` and change the legend with the `leg` keyword:

```julia
plot(sol,plot_analytic=true,w=[3 2],leg=false)
```

You can do all sorts of things. Have fun!
